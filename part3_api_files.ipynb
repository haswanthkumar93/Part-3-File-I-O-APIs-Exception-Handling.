```python
import requests

BASE_URL = "https://dummyjson.com/products"

# ----------------------
# Step 1: Fetch 20 Products
# ----------------------
print("Fetching Products...\n")

resp = requests.get(f"{BASE_URL}?limit=20")
resp.raise_for_status()  # will raise if non-200
data = resp.json()

products = data.get("products", [])

print(f"{'ID':<3} | {'Title':<30} | {'Category':<15} | {'Price':<8} | Rating")
print("-" * 80)

for p in products:
    print(f"{p['id']:<3} | {p['title']:<30} | {p['category']:<15} | ${p['price']:<7.2f} | {p['rating']}")

# ----------------------
# Step 2: Filter & Sort
# ----------------------
print("\nFiltered Products (Rating >= 4.5):\n")

filtered = [p for p in products if p.get("rating", 0) >= 4.5]
filtered.sort(key=lambda x: x.get("price", 0), reverse=True)

for p in filtered:
    print(f"{p['title']:<30} | Price: ${p['price']:.2f} | Rating: {p['rating']}")

# ----------------------
# Step 3: Fetch Laptops
# ----------------------
print("\nLaptops Category:\n")

resp = requests.get(f"{BASE_URL}/category/laptops")
resp.raise_for_status()
laptop_data = resp.json()
laptops = laptop_data.get("products", [])

for l in laptops:
    print(f"{l['title']:<30} | Price: ${l['price']:.2f}")

# ----------------------
# Step 4: POST Request
# ----------------------
print("\nSending POST Request...\n")

new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API",
}

resp = requests.post(f"{BASE_URL}/add", json=new_product)
resp.raise_for_status()

print("Response from server:")
print(resp.json())
```

```python
import requests

BASE_URL = "https://dummyjson.com/products"

# ----------------------
# Part A: Safe Divide
# ----------------------
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

print("Safe Divide Tests:")
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))


# ----------------------
# Part B: Safe File Reader
# ----------------------
def read_file_safe(filename):
    try:
        with open(filename, "r") as file:
            return file.read()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")

print("\nReading python_notes.txt:")
print(read_file_safe("python_notes.txt"))

print("\nReading ghost_file.txt:")
print(read_file_safe("ghost_file.txt"))


# ----------------------
# Part C: Robust API Calls
# ----------------------
print("\nFetching Products with Error Handling...\n")

try:
    resp = requests.get(f"{BASE_URL}?limit=20", timeout=5)
    resp.raise_for_status()
    data = resp.json()

    products = data.get("products", [])

    for p in products:
        print(f"{p['title']} - ${p['price']}")

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print("Error:", e)


# ----------------------
# Part D: Input Validation Loop
# ----------------------
while True:
    user_input = input("\nEnter a product ID to look up (1–100), or 'quit' to exit: ")

    if user_input.lower() == "quit":
        print("Exiting program.")
        break

    # Validate number
    if not user_input.isdigit():
        print("Invalid input. Please enter a number.")
        continue

    product_id = int(user_input)

    # Validate range
    if product_id < 1 or product_id > 100:
        print("Please enter a number between 1 and 100.")
        continue

    # API call with error handling
    try:
        resp = requests.get(f"{BASE_URL}/{product_id}", timeout=5)

        if resp.status_code == 404:
            print("Product not found.")
        else:
            resp.raise_for_status()
            product = resp.json()
            print(f"{product['title']} - ${product['price']}")

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)
```

```python
import requests
from datetime import datetime

BASE_URL = "https://dummyjson.com/products"
LOG_FILE = "error_log.txt"

# ----------------------
# Logger Function
# ----------------------
def log_error(message):
    with open(LOG_FILE, "a") as f:
        timestamp = datetime.now()
        f.write(f"[{timestamp}] {message}\n")


# ----------------------
# Trigger ConnectionError
# ----------------------
print("Triggering ConnectionError...\n")

try:
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
except requests.exceptions.ConnectionError as e:
    print("Connection failed. Please check your internet.")
    log_error(f"ERROR in fetch_products: ConnectionError — {e}")


# ----------------------
# Trigger HTTP 404 Error
# ----------------------
print("\nTriggering HTTP 404...\n")

try:
    resp = requests.get(f"{BASE_URL}/999", timeout=5)

    if resp.status_code != 200:
        print("Product not found.")
        log_error("ERROR in lookup_product: HTTPError — 404 Not Found for product ID 999")
    else:
        resp.raise_for_status()

except requests.exceptions.ConnectionError as e:
    print("Connection failed. Please check your internet.")
    log_error(f"ERROR in lookup_product: ConnectionError — {e}")

except requests.exceptions.Timeout as e:
    print("Request timed out. Try again later.")
    log_error(f"ERROR in lookup_product: Timeout — {e}")

except Exception as e:
    print("Error:", e)
    log_error(f"ERROR in lookup_product: {e}")


# ----------------------
# Display Log File
# ----------------------
print("\nError Log Contents:\n")

try:
    with open(LOG_FILE, "r") as f:
        print(f.read())
except FileNotFoundError:
    print("No log file found.")
```
